```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df1= pd.read_csv('downloads/winequality-red.csv', sep=';')
```

```python
df1
```

```python
df2= pd.read_csv('downloads/winequality-white.csv', sep=';')
```

```python
df2
```

```python
df1.columns==df2.columns
```

```python
df1['colour']= np.repeat('red', df1.shape[0])
```

```python
df1
```

```python
df2['colour']= np.repeat('white', df2.shape[0])
```

```python
df2.head()
```

```python
df= pd.concat([df1,df2])
```

```python
df
```

```python
import seaborn as sns
```

```python
df.columns
```

# Histograms for Various Features

```python
for cols in df.columns:
    plt.figure()
    plt.title(f'{cols}')
    df[cols].hist()
```

# Scatterplots of Quality Against Various Features

```python
for cols in df.columns:
    plt.figure()
    plt.title(f'{cols}')
    sns.scatterplot(x='quality', y= df[cols], data= df)
```

# Drawing Conclusions Using Groupby

## Is a certain type of wine associated with higher quality?

```python
df.head()
```

```python
df.groupby(by='colour')['quality'].agg(['mean'])
```

## What level of acidity receives the highest average rating?

```python
# View the min, 25%, 50%, 75%, max pH values with Pandas describe
df['pH'].describe()
```

```python
df['acidity_levels']= pd.cut(x=df['pH'], bins=[2.72,3.11, 3.21, 3.32, 4.01], labels=['high', 'mod_high', 'medium', 'low'])
```

```python
df.head()
```

```python
df.groupby(by='acidity_levels')['quality'].agg(['mean'])
```

## Do wines with higher alcoholic content receive better ratings?

```python
df.groupby(by='alcohol')['quality'].agg(['mean'])
```

```python
df['alcohol'].median()
```

```python
df['alcohol'].describe()
```

```python
df['alcohol_levels']= pd.cut(x=df['alcohol'], bins=[0, 10.3, 14.9], labels=['low', 'high'])
```

```python
df.groupby(by='alcohol_levels')['quality'].agg(['mean'])
```
