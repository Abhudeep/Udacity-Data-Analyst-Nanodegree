```python
# doing ANOVA test to check whether the average response time across complaint types is similar or not
# h0= average response time across complaint types is similar if p>0.05
# ha=average response time across complaint types is not similar so for proove that we have cheake p value p<0.05
```

```python
anova_df=pd.DataFrame()
anova_df["Elapsed_Time"]=df1["Elapsed Time"]
anova_df["Complaint_Type"]=df1["Complaint Type"]

anova_df.dropna(inplace=True)
anova_df.head()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

lm=ols("Elapsed_Time~Complaint_Type",data=anova_df).fit()
table=sm.stats.anova_lm(lm)
table
```
